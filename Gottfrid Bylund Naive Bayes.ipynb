```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
```

```python
#läser in data, men lägger även till en kolumn; risk (vår Y variabel)
df = pd.read_csv('credit_data2.csv')
```

```python
#ta bort NaN värden
df.dropna()
```

```python
#skapa dummyvariabler för att kunna predikta outcome
# vi använder oss av följande kolumner;
data = pd.get_dummies(df[['Sex', 'Housing', 'Saving accounts', 'Checking account', 'Purpose','Risk"']])
```

```python
data
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
#vill använda Bayes för att se om datan kan predikta dem med dålig kreditrisk, droppa därför den i
#x variabeln men behåller i y variabeln
x = data.drop('Risk"_bad"', axis=1).values
y = data['Risk"_bad"'].values
```

```python
y
```

```python
#splittar data till träning och test dataset. =.3 betyder att vi delar datan till 70% träning och 30% test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
seed = 5
```

```python
#dags att testa vår modell och se hur bra den är
model = GaussianNB()
train_model = model.fit(x_train, y_train)
y_model = train_model.predict(x_train)

print("totalt antal felaktiga av totalt %d punkter, med %d fel" % (x_train.shape[0],(y_train != y_model).sum()))
```

```python
#perfekt score, är modellen overfit? Ska testa detta mot X testsetet;

y_pred = train_model.predict(x_test)
print("totalt antal felaktiga av totalt %d punkter, med %d fel" % (x_train.shape[0],(y_train != y_pred)))
```
